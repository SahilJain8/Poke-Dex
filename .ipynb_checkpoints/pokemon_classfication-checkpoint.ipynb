```python
from datasets import load_dataset
import  os
pokemon_image_datapath = "pokemon_data/archive (1)/PokemonData"

dataset = load_dataset(pokemon_image_datapath)
```

```python
dataset_split = dataset["train"].train_test_split(test_size=0.2,shuffle=True)
```

```python
train_dataset = dataset_split["train"]
test_dataset = dataset_split["test"]
```

```python
import os
class_name = os.listdir(pokemon_image_datapath)
```

```python
#Display the images
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random as rnd
from PIL import Image
```

```python
def display_image(image:Image,label:str) -> None:
    imageArray = np.asarray(image)
    plt.imshow(imageArray)
    plt.title(class_name[label])
```

```python
index = rnd.randint(0,len(train_dataset))

display_image(train_dataset[index]['image'],train_dataset[index]['label'])
```
